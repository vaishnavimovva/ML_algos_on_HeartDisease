```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the bars
categories = ['XGBoost', 'RandomForest', 'NaviesBayes','SVM', 'Logistic Regression','LSTM']
values1 = [100, 90.38, 90.72, 75.78, 90.0, 92.65]
values2 = [100, 87.85, 82.24, 67.28, 84.11, 90.67]
values3 = [100, 87.5, 84.78, 68.47, 85.32, 90.48]
values4 = [100, 89.09, 86.27, 71.28,86.95,91.72]

# Combine the data into a single DataFrame
data = pd.DataFrame({
    'Categories': categories * 4,
    'Values': values1 + values2 + values3 + values4,
    'Graph': ['Precision'] * len(categories) + ['Recall'] * len(categories) + ['Accuracy'] * len(categories) + ['F1_Score'] * len(categories)
})

# Set the style for the plot (optional)
sns.set_style("whitegrid")

# Create the bar graph using seaborn
plt.figure(figsize=(10, 6))  # Set the figure size (width=10 inches, height=6 inches)
sns.barplot(x='Categories', y='Values', hue='Graph', data=data, palette=['blue', 'green', 'orange', 'red'],edgecolor='black')

# Add labels and title
plt.xlabel('Models')
plt.ylabel('Performance')
plt.title('Various machine learning Algorithms and their performances')

# Set y-axis limits and ticks
plt.ylim(65, 105)
plt.yticks(np.arange(60, 105, 3))

# Show the plot
plt.legend(title='Factors', loc='upper right')
plt.tight_layout()
plt.show()
```
